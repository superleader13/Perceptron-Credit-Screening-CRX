```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.cluster import KMeans
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
from sklearn.metrics import davies_bouldin_score
```

```python
df= pd.read_csv('crx.data') 
df.head(5)
```

```python
df.dtypes
#cek tipe data
```

```python
#cek missing value ada atau tidak 
print(df.isnull().sum())
```

```python
#mengatasi missing value untuk type data numerik (float,int) dengan mean 
df['A2'] = df['A2'].fillna((df['A2'].mean()))
df['A3'] = df['A3'].fillna((df['A3'].mean()))
df['A14'] = df['A14'].fillna((df['A14'].mean()))
```

```python
#mengatasi missing value untuk type data object
for col in df.columns:
    
    #untuk mengisi missing value pada type data "object" dengan modus
    if df[col].dtype == 'object':                    
        df = df.fillna(max(df[col].value_counts()))
        
print(df.isnull().sum()) 
```

```python
#merubah tpe data object menjadi string 
le=LabelEncoder()

for col in df.columns.to_numpy():
    if df[col].dtypes == 'object':
        df[col] = le.fit_transform(df[col].astype(str))
df.head(5)
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters = 5, random_state=0, init="random")
kmeans.fit(df)                                         #melatih data df dengan algoritma kmeans
kmeans.labels_
```

```python
#memprediksi data
predict =kmeans.predict(df)
predict
```

```python
#membuat kolom "cluster" yang berisikan hasil predict
df['cluster']=predict
df.head(5)
```

```python
db_index = davies_bouldin_score(df, predict)
print(db_index)
```
